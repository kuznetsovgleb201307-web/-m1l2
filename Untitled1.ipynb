```python
def game():
    guest = int(input('Hi! I am Bla Bla Bot! Decide this! 2 + 2 = ?'))
    rightwrong = ''
    if guest == 4:
        rightwrong = 'You are right!'
    else:
        rightwrong = 'Oh no! Wrong'
    return rightwrong
```

```python
import random

def gen_pass(pass_length):
    elements = "+-/*!&$#?=@<>123456789"
    password = ""
    for i in range(pass_length):
        password += random.choice(elements)
    return password
```

```python
import telebot
bot = telebot.TeleBot("")

@bot.message_handler(commands=['start'])
def send_welcome(message):
    bot.reply_to(message, "Привет! Я твой Telegram бот. Напиши что-нибудь!Напривер /list!")

@bot.message_handler(commands=['hello'])
def send_hello(message):
    bot.reply_to(message, "Привет! Как дела?")

@bot.message_handler(commands=['bye'])
def send_bye(message):
    bot.reply_to(message, "Пока! Удачи!")

@bot.message_handler(commands=['password'])
def send_bye(message):
    bot.reply_to(message, gen_pass(12))


@bot.message_handler(commands=['game'])
def send_bye(message):
    bot.reply_to(message, "2+2")

@bot.message_handler(commands=['4'])
def send_bye(message):
    bot.reply_to(message, "правильно!")

@bot.message_handler(commands=['game2'])
def send_bye(message):
    bot.reply_to(message, "2+44")

@bot.message_handler(commands=['46'])
def send_bye(message):
    bot.reply_to(message, "правильно!")


@bot.message_handler(commands=['list'])
def send_welcome(message):
    bot.reply_to(message, '''
/start    - перезапустить бота
/hello    - сказать привет
/bye      - сказать пока
/password - сгенирировать рандомный пароль из 12 символов
/game     - сыграть в игру
/game2    - второй раунд''')

@bot.message_handler(func=lambda message: True)
def echo_all(message):
    bot.reply_to(message, message.text)

bot.polling()
```
