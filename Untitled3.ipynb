```python
import telebot
import os
import random

bot = telebot.TeleBot('TOKEN')

@bot.message_handler(commands=['start'])
def send_welcome(message):
    bot.reply_to(message, "Здравствуйте! Я ваш бот - помощник природы! Напиши что-нибудь ! Например /list!")

@bot.message_handler(commands=['hello'])
def send_hello(message):
    bot.reply_to(message, "Здравствуйте! Как дела?")

@bot.message_handler(commands=['bye'])
def send_bye(message):
    bot.reply_to(message, "Досвидания! Удачи!")

@bot.message_handler(commands=['fakt'])
def send_mem(message):
    fakti = random.choice(os.listdir('fakts'))
    with open(f'fakts/{fakti}', 'rb') as f:
        bot.send_photo(message.chat.id, f)

@bot.message_handler(commands=['list'])
def send_bye(message):
    bot.reply_to(message, '''
/start        - перезапустить бота
/hello        - сказать здравствуйте
/bye          - сказать досвидания
/fakt         - рандомный факт
/secret       - секрет
/fakttext1    - факт (текстовый)
/fakttext2    - факт второй (текстовый)
    ''')


@bot.message_handler(commands=['fakttext1'])
def send_bye(message):
    bot.reply_to(message, "Ежегодно в мировой океан попадает около 8 миллионов тонн пластика. За последние 50 лет количество пластиковых отходов увеличилось более чем в 20 раз.")

@bot.message_handler(commands=['fakttext2'])
def send_hello(message):
    bot.reply_to(message, "Ежедневно на Земле исчезает до 100 видов растений и животных. Популяции позвоночных животных сократились на 68% с 1970 года. ")


@bot.message_handler(commands=['secret'])
def send_mem(message):
    with open('secret/mem7.png', 'rb') as f:
        bot.send_photo(message.chat.id, f)

bot.polling()

```
